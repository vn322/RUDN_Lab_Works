## 5) Уравнение f = 0, левая часть f которого является много-
членом первой степени кольца A[x1, . . . , xn], называется линейным уравне-
нием над кольцом A с n неизвестными x1, .. . ,xn или линейным уравнением
из кольца A[x1, . . . , xn].
Более развернуто, линейное уравнение - это уравнение вида
a1x1 + · · · + anxn = b,
где коэффициенты a1, . . . , an и b принадлежат кольцу A и среди a1, . . . , an
имеются отличные от нуля. 

линейное уравнение принадлежит кольца A[x1, . . . , xn], подразумевая, что
это уравнение можно привести к виду f = 0, f ∈ A[x1, . . . , xn].


Конечное множество уравнений из кольца A[x1, . . . , xn]
будет системой уравнений.

##  7) Всякая линейная система над целостным кольцом или приводится к треугольному виду, или не имеет решения ни в одном из расширений этого кольца.
 Пусть S - заданная система уравнений из A[x1, . . . , xn], L, T - пока пустые множества
уравнений. Будем преобразовывать систему  так, чтобы
исходная система была на каждом шаге эквивалентна объединению S, а в
конце S оказалось пустым, а T - треугольной системой. L будет вспомогательным множеством.
Один шаг опишем так. Пусть T +S эквивалентно исходной системе и T
пусто или
lm(f) < lm(t) ∀f ∈ S, t ∈ T.
Прежде всего найдем
m = max
f∈S
lm(f)
и положим
L = {f ∈ S : lm(f) = m}, S = {f ∈ S : lm(f) < m}
Тогда исходной системе эквивалентно система T + L + S. Пусть для определенности
L = (g1, . . . , gk).
Добавим первый элемент g1 множества L к T, что не меняет треугольной
формы этого множества, поскольку
m < lm(t) ∀t ∈ T.

Переберем теперь все оставшиеся g из L. Поскольку
gi = lc(gi)m + . . .
разность
gi" = lc(g1)gi − lc(gi)g1
не содержит монома m, т.е.
lm(gi") < m.
Если степень gi" равна −1, то этот многочлен сводится к 0, что дает тривиальное уравнение 0 = 0, которое мы просто отбросим. Если степень gi"
равна 0, то это дает уравнение вида 1 = 0, которое не может иметь решения,
поскольку A - целостное кольцо. Если же степень равна 1, то получается
линейное уравнение. Добавим его к списку S. Сделав это со всеми g из L,
мы получим два множества T != ∅ и S, причем T + S будет эквивалентна
исходной и
lm(f) < m ≤ lm(t) ∀f ∈ S, t ∈ T.
Делая такие шаги мы будем каждый раз увеличивать T и уменьшать S
хотя бы на один элемент. Поэтому за конечное число шагов мы придем к
S = ∅ и тем самым приведем систему к треугольному виду.
Описанный алгоритм приведения произвольной системы к системе треугольного вида часто называют алгоритмом Гаусса.

уравнений входит в треугольную систему?


```python
def triangulation(S):
    T=[]
    n=0
    while S!=[]:
        m=max([s.lm() for s in S])
        L = [s for s in S if s.lm() == m]
        S = [s for s in S if s.lm() < m]
        g1=L[0]
        T.append(g1)
        for g in L[1:]:
            g=g1.lc()*g-g.lc()*g1
            if g.degree()==0:
                T.append(g)
                break
            elif g.degree()==1:
                S.append(g)
    return T
```
